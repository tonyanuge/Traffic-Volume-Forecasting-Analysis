```python
# Example: T-test to compare traffic volume in different halves of the year
first_half = df_Stats[df_Stats['Week'] <= 26]['VALUE']
second_half = df_Stats[df_Stats['Week'] > 26]['VALUE']

t_stat, p_value = stats.ttest_ind(first_half, second_half)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

Hypothesis Probability Question: "Is there a significant difference in the average weekly volume of cars between the first half and the second half of the year 2019 in Dublin, Ireland?"

Rationale: This hypothesis aims to determine if traffic volumes significantly differ between the two halves of the year, suggesting seasonal variation in traffic patterns.

The T-test result, with a T-statistic of -1.7032 and a P-value of 0.0891, is used to assess this hypothesis. The negative T-statistic indicates that the average traffic volume in the first half is lower than in the second half. However, the P-value is above the conventional threshold of 0.05, which suggests that this difference is not statistically significant.

Decision on Hypothesis: We fail to reject the null hypothesis because the P-value is greater than 0.05. This implies that there isn't enough statistical evidence to conclude a significant difference in traffic volumes between the first and second halves of the year. The observed difference might be due to random variations rather than a consistent seasonal effect.

# Example using a Chi-Square Test, assuming "Traffic Site" has two categories - Urban and Regional

# Sample Selection, Aggregation, and ANOVA Analysis on Traffic Data Across Dublin, Leeds, and Edinburgh

```python
import numpy as np
import statsmodels.api as sm  # Importing statsmodels
from statsmodels.formula.api import ols  # Importing the ols function
```

```python
# Determining sample size - a common rule of thumb is to use a sample size that is 10% of the population,
# but with a minimum of 30 and a maximum of 500 for larger datasets.

sample_size_dublin = min(500, max(30, int(0.1 * len(df_Dublin))))
sample_size_leeds = min(500, max(30, int(0.1 * len(df_Leeds))))
sample_size_edinburgh = min(500, max(30, int(0.1 * len(df_Edinburgh))))
```

```python
# Generating a random sample from each dataset
np.random.seed(0)  # For reproducibility
df_Dublin_sample = df_Dublin.sample(n=sample_size_dublin)
df_Leeds_sample = df_Leeds.sample(n=sample_size_leeds)
df_Edinburgh_sample = df_Edinburgh.sample(n=sample_size_edinburgh)
```

```python
# Aggregating the sampled data yearly
df_Dublin_sample_yearly = df_Dublin_sample.groupby('Year')['VALUE'].mean().reset_index()
df_Dublin_sample_yearly.rename(columns={'VALUE': 'Average_weekly_volume_of_cars'}, inplace=True)

df_Leeds_sample_yearly = df_Leeds_sample[['year', 'all_motor_vehicles']].groupby('year').mean().reset_index()
df_Edinburgh_sample_yearly = df_Edinburgh_sample[['year', 'all_motor_vehicles']].groupby('year').mean().reset_index()


```

```python
# Preparing data for ANOVA with sampled dataset
anova_data_sampled = pd.concat([
    df_Dublin_sample_yearly.assign(City='Dublin'),
    df_Leeds_sample_yearly.assign(City='Leeds', Average_weekly_volume_of_cars=df_Leeds_sample_yearly['all_motor_vehicles']),
    df_Edinburgh_sample_yearly.assign(City='Edinburgh', Average_weekly_volume_of_cars=df_Edinburgh_sample_yearly['all_motor_vehicles'])
])

# Performing ANOVA on the sampled data
anova_model_sampled = ols('Average_weekly_volume_of_cars ~ C(City)', data=anova_data_sampled).fit()
anova_results_sampled = sm.stats.anova_lm(anova_model_sampled, typ=2)

anova_results_sampled
```

The ANOVA analysis was conducted on the sampled data from each city's dataset, and the results are as follows:

A higher F-statistic value generally indicates a greater degree of difference between the group means. In your case, an F-statistic of approximately 600.01 is quite high, suggesting that the average weekly volume of cars differs significantly across the three cities.

P-Value (1.097646e-34): Under the null hypothesis, which in an ANOVA is usually the hypothesis that all group means are equal, the p-value represents the likelihood of seeing the data or something more extreme. A very low p-value indicates that, in accordance with the null hypothesis, the observed data is extremely unlikely. The p-value in your analysis is incredibly low—nearly zero—and is much below any typical significance threshold (such as 0.05 or 0.01). This indicates that there is almost no chance of finding such a high F-statistic under the null hypothesis, which leads you to reject it.

Degrees of Freedom: The numerator (dfn) in your analysis has two degrees of freedom, while the denominator (dfd) has forty-eight. The degrees of freedom for the denominator are usually the total number of observations minus the number of groups, and for the numerator, they are usually the number of groups minus one (3 cities - 1 = 2). The F-distribution's shape is ascertained using the degrees of freedom, and the p-value is subsequently computed using this information.

In conclusion, the findings clearly imply that the average weekly volume of cars in each of the three cities varies significantly. These differences are statistically significant and unlikely to have happened by accident, according to the incredibly low p-value.



The plot displays your ANOVA analysis's findings. The F-distribution with the designated degrees of freedom is represented by the blue curve. # Statistical Comparison of Average Weekly Vehicle Volumes Between Cities Using T-Test, Wilcoxon, and Mann-Whitney U Test

```python
from scipy.stats import ttest_ind, wilcoxon, mannwhitneyu

# Pairwise comparisons for T-Test, Wilcoxon, and Mann-Whitney Tests
pairs = [("Dublin", "Leeds"), ("Dublin", "Edinburgh"), ("Leeds", "Edinburgh")]

# Initializing dictionaries to store test results
ttest_results = {}
wilcoxon_results = {}
mannwhitney_results = {}

# Conducting the tests
for city1, city2 in pairs:
    # Extracting data for each pair
    data1 = anova_data_sampled[anova_data_sampled['City'] == city1]['Average_weekly_volume_of_cars']
    data2 = anova_data_sampled[anova_data_sampled['City'] == city2]['Average_weekly_volume_of_cars']

    # T-Test
    ttest_results[f"{city1} vs {city2}"] = ttest_ind(data1, data2, equal_var=False)

    # Wilcoxon Test - requires equal lengths of data, hence we take a sample of equal size from each
    min_size = min(len(data1), len(data2))
    wilcoxon_results[f"{city1} vs {city2}"] = wilcoxon(data1.sample(min_size), data2.sample(min_size))

    # Mann-Whitney Test
    mannwhitney_results[f"{city1} vs {city2}"] = mannwhitneyu(data1, data2)

# Displaying the results
ttest_results, wilcoxon_results, mannwhitney_results

```

T-Test: To compare the means between two groups. This test will be applied pairwise (Dublin vs Leeds, Dublin vs Edinburgh, Leeds vs Edinburgh).

Wilcoxon Test: A non-parametric test that compares two paired groups. I'll use this test similarly to the T-Test for pairwise comparisons.

Mann-Whitney Test: Another non-parametric test, used to compare differences between two independent groups. I'll also apply this test pairwise.

After performing these tests, i'll create visualizations to illustrate the findings and provide evidence to support the methodology.

Let's start with the statistical tests and then proceed to the visualizations.

The results of the T-Test, Wilcoxon Test, and Mann-Whitney Test for pairwise comparisons between Dublin, Leeds, and Edinburgh are as follows:

# Creating a Synthetic School Calendar Dataset with Holiday and School Day Identification

```python
import pandas as pd
from datetime import datetime, timedelta
import warnings
# Ignoring all warnings
warnings.filterwarnings('ignore')

# Create a function to generate a synthetic school calendar dataset
def create_school_calendar(start_year, end_year):
    # Define typical school year start and end dates
    school_start_month, school_end_month = 9, 6  # September and June
    holiday_dates = {
        'Christmas': {'start': 12, 'end': 1, 'start_day': 24, 'end_day': 7},  # Christmas break
        'Easter': {'start': 3, 'end': 4, 'start_day': 15, 'end_day': 30},  # Easter break
        'Midterm_1': {'start': 10, 'end': 10, 'start_day': 28, 'end_day': 31},  # October break
        'Midterm_2': {'start': 2, 'end': 2, 'start_day': 14, 'end_day': 18},  # February break
    }

    # Initialize an empty DataFrame
    calendar_df = pd.DataFrame(columns=['Year', 'Date', 'Is_School_Day'])

    # Generate data for each year in the range
    for year in range(start_year, end_year + 1):
        # Generate dates for the whole year
        start_date = datetime(year, 1, 1)
        end_date = datetime(year, 12, 31)
        total_days = (end_date - start_date).days + 1
        dates = [start_date + timedelta(days=x) for x in range(total_days)]

        # Determine if each date is a school day
        for date in dates:
            is_school_day = True
            if date.month < school_start_month and date.month > school_end_month:
                is_school_day = False  # Summer break
            for holiday_name, holiday_info in holiday_dates.items():
                if date.month == holiday_info['start'] and date.day >= holiday_info['start_day']:
                    is_school_day = False
                if date.month == holiday_info['end'] and date.day <= holiday_info['end_day']:
                    is_school_day = False

            calendar_df = calendar_df.append({'Year': year, 'Date': date, 'Is_School_Day': is_school_day}, ignore_index=True)

    return calendar_df

```

```python

from sklearn.utils import shuffle
import numpy as np

# Function to generate new data based on existing patterns
def generate_new_data(df, num_rows):
    new_data = pd.DataFrame()

    # Handling each column
    for column in df.columns:
        # If the column is categorical or datetime, we sample from existing values
        if df[column].dtype == 'object' or column in ['Date', 'Weeks of the Year']:
            new_data[column] = np.random.choice(df[column].unique(), size=num_rows)
        # For numeric columns, we generate data within the range of existing values
        elif df[column].dtype in ['float64', 'int64']:
            min_val, max_val = df[column].min(), df[column].max()
            new_data[column] = np.random.uniform(min_val, max_val, size=num_rows)
        else:
            new_data[column] = np.random.choice(df[column].unique(), size=num_rows)

    return new_data


```

```python
# Function to clean and remove stopwords from a review
def clean_review(review):
    # Remove non-alphabetic characters and lowercase the words
    review = re.sub("[^a-zA-Z]", " ", review).lower()

    # Tokenize the review
    word_tokens = word_tokenize(review)

    # Filter out stopwords
    filtered_review = [word for word in word_tokens if word not in stop_words]

    # Rejoin words to form the cleaned review
    return ' '.join(filtered_review)
```

```python
import spacy

# Load the spaCy model (make sure to install it first if not already done)
try:
    nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])  # Disable parser and NER for efficiency
except Exception as e:
    print(f"Error loading spaCy model: {e}")
```

```python
# Function to clean and remove stopwords using spaCy
def clean_review_spacy(review):
    if not review:
        return ""

    # Process the review text
    doc = nlp(review.lower())

    # Filter out stopwords and non-alphabetic characters
    filtered_review = [token.text for token in doc if token.text.isalpha() and not token.is_stop]

    # Rejoin words to form the cleaned review
    return ' '.join(filtered_review)
```

# BAG OF WORDS

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
import re
import nltk
```

```python
# Download necessary NLTK data
#nltk.download('punkt')
#nltk.download('stopwords')

# Split the content into individual reviews
Transport_reviews = file_content_without_video_ids.split('\n')
```

```python
# Function to clean the text
def clean_text(text):
    text = re.sub("[^a-zA-Z]", " ", text)  # Remove non-letters
    text = text.lower()  # Convert to lower case
    words = nltk.word_tokenize(text)  # Tokenize into words
    words = [w for w in words if w not in stopwords.words('english')]  # Remove stopwords
    return " ".join(words)

# Clean the reviews
cleaned_reviews = [clean_text(review) for review in Transport_reviews]

# Create a Bag of Words model
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(cleaned_reviews)

# X is the matrix representing the Bag of Words model of the revie
```

```python
# Convert the Bag of Words matrix to a DataFrame
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Print the DataFrame
print(bow_df.head())
```

Explanation of Output:

The number of times a specific word column appears in a given review row is indicated in each cell. The majority of values are 0, meaning that word is not present in that review.
Many words do not appear in every document in BoW models, which is typical of this sparse representation.
Justification:
Using the BoW model method enables algorithms to find patterns or trends in the text, like recurring words or phrases, by turning text into numerical data.
